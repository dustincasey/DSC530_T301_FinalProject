```python
# Running a normal distribution for point margin.

def MakeNormalPlot(margin):
    mean = margin.mean()
    std = margin.std()
    xs = [-4, 4]
    fxs, fys = thinkstats2.FitLine(xs, inter=mean, slope=std)
    thinkplot.Plot(fxs, fys, color='gray', label='model')
    xs, ys = thinkstats2.NormalProbability(margin)
    thinkplot.Plot(xs, ys, label='points margin')

MakeNormalPlot(regular_season.points_margin)

# I believe a normal distribution fits best for the points margin variable because from the histogram, it appeared to have a
# normal distribution. After running a Normal Probability Plot (NPP), the normal distribution line aligns well with the
# modelled distribution. Although the lines deviate slightly after 2 standard deviations, I believe the 
# fit is still the best possible.
```

```python
import numpy as np
import math

# Creating lines of best fit for scatter plots.
denominator_1 = regular_season.points_game.dot(regular_season.points_game)-regular_season.points_game.mean()*regular_season.points_game.sum()
m_1 = (regular_season.points_game.dot(regular_season.points_margin)-regular_season.points_margin.mean()*regular_season.points_game.sum())/denominator_1
b_1 = (regular_season.points_margin.mean()*regular_season.points_game.dot(regular_season.points_game)-regular_season.points_game.mean()*regular_season.points_game.dot(regular_season.points_margin))/denominator_1
y_pred_1 = m_1*regular_season.points_game + b_1

denominator_2 = regular_season.opp_points_game.dot(regular_season.opp_points_game)-regular_season.opp_points_game.mean()*regular_season.opp_points_game.sum()
m_2 = (regular_season.opp_points_game.dot(regular_season.points_margin)-regular_season.points_margin.mean()*regular_season.opp_points_game.sum())/denominator_2
b_2 = (regular_season.points_margin.mean()*regular_season.opp_points_game.dot(regular_season.opp_points_game)-regular_season.opp_points_game.mean()*regular_season.opp_points_game.dot(regular_season.points_margin))/denominator_2
y_pred_2 = m_2*regular_season.opp_points_game + b_2

# Creating scatter plots for points allowed vs point margin
thinkplot.PrePlot(2, cols=2)
thinkplot.Scatter(regular_season.points_game,regular_season.points_margin)
thinkplot.Plot(regular_season.points_game,y_pred_1,color="grey")
thinkplot.Config(xlabel='Points Scored per Game',
              ylabel ='Point Margin')

thinkplot.PrePlot(2)
thinkplot.SubPlot(2)
thinkplot.Scatter(regular_season.opp_points_game,regular_season.points_margin)
thinkplot.Plot(regular_season.opp_points_game,y_pred_2,color="grey")
thinkplot.Show(xlabel='Points Allowed per Game',
              ylabel ='Point Margin')

# Running correlation between the two variables.
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)
    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    
    meanx = np.mean(xs)
    varx = np.var(xs)
    meany = np.mean(ys)
    vary = np.var(ys)
    
    corr = Cov(xs, ys, meanx, meany) / math.sqrt(varx * vary)
    return corr

def SpearmanCorr(xs, ys):
    xranks = pd.Series(xs).rank()
    yranks = pd.Series(ys).rank()
    return Corr(xranks, yranks)

print("Correlation tests between Points Scored per Game and Point Margin")
print("Covariance: \t",Cov(regular_season.points_game,regular_season.points_margin))
print("Pearson's Corr: ",Corr(regular_season.points_game,regular_season.points_margin))
print("Spearmans's Corr: ",SpearmanCorr(regular_season.points_game,regular_season.points_margin))

print("\nCorrelation tests between Points Allowed per Game and Point Margin")
print("Covariance: \t",Cov(regular_season.opp_points_game,regular_season.points_margin))
print("Pearson's Corr: ",Corr(regular_season.opp_points_game,regular_season.points_margin))
print("Spearmans's Corr: ",SpearmanCorr(regular_season.opp_points_game,regular_season.points_margin))
```

```python
import random
class HypothesisTest(object):
    def __init__(self, data):
        self.data = data
        self.MakeModel()
        self.actual = self.TestStatistic(data)
        
    def PValue(self, iters=1000):
        self.test_stats = [self.TestStatistic(self.RunModel())
                            for _ in range(iters)]
        
        count = sum(1 for x in self.test_stats if x >= self.actual)
        return count / iters
    
    def TestStatistic(self, data):
        raise UnimplementedMethodException()
        
    def MakeModel(self):
        pass
    def RunModel(self):
        raise UnimplementedMethodException()
        
class DiffMeansPermute(thinkstats2.HypothesisTest):
    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat
    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))
    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
    
# Creating two datasets, Big 10 conference teams and all others.
big_ten = regular_season[regular_season.conf_name == "Big Ten"]
non_big_ten = regular_season[regular_season.conf_name != "Big Ten"]

# Feeding in data for wins by conference.
data = big_ten.win.values,non_big_ten.win.values
ht = DiffMeansPermute(data)

# Calculating a P-value.
pvalue = ht.PValue()
print(pvalue)
print(ht.actual,ht.MaxTestStat())

ht.PlotCdf()
thinkplot.Config(xlabel='test statistic',
                   ylabel='CDF')
```
